```python
from fastai import *
from fastai.vision import *
```

```python
# Downloading images from the csv file of URL's
path = Path('gadgets')

for file, folder in [('Airpods.csv', 'Airpods'), ('AppleWatch.csv', 'AppleWatch'), 
                     ('Iphone.csv', 'Iphone'), ('MacbookPro.csv', 'MacbookPro')]:
    dest = path/folder # path + '/' + folder
    dest.mkdir(parents=True, exist_ok=True)
    download_images(path/file, dest)
```

```python
# Verifying for corrupt images
for folder in ('Airpods', 'AppleWatch', 'Iphone', 'MacbookPro'):
    print(folder)
    verify_images(path/folder, delete=True, max_size=500)
```

```python
# Viewing our data 
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train='.', valid_pct=0.2,
                                  ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
data.classes
```

```python
data.show_batch(rows=4, figsize=(12,10))
```

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```python
# Creating a model
from fastai.metrics import error_rate
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
defaults.device = torch.device('cuda')
learn.fit_one_cycle(4)
```

```python
 learn.save("stage-1")
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(4, max_lr=slice(3e-5,3e-4))
```

```python
learn.save('stage-2')
```

```python
learn.load('stage-2');
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, figsize=(12,12))
```
